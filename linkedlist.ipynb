```python
class node:
    def __init__(self,value):
        self.data=value
        self.next=None

class linkedlist:
    def __init__(self):
        self.head=None
    def insert(self,value):
        new=node(value)
        if self.head==None:
            self.head=new
        else:
            new.next=self.head
            self.head=new
    def __str__(self):
        curr=self.head
        re=''
        while curr:
            re+=str(curr.data)+'->'
            curr=curr.next
        return re[:-2]
    def append(self,value):
        new=node(value)
        if self.head==None:
            self.head=new
        else:
            curr=self.head
            while curr.next:
                curr=curr.next
            curr.next=new
    def delete(self,target):
        curr=self.head
        while curr:
            if curr.next.data==target:
                break
            curr=curr.next
        a=curr.next
        curr.next=curr.next.next
        a.next=None
```

```python
# use linkedlist and node classes defined in a previous cell

ll = linkedlist()
print("empty list:", ll)

# insert at head
ll.insert(10)
ll.insert(20)
ll.insert(30)
print("after inserts:", ll)

# append at tail
ll.append(40)
print("after append 40:", ll)

# delete a middle element (safe with current delete implementation)
```
